## missing data processing

```python
# Removed three irrelevant columns: 'id', 'host_name', 'last_review'
airbnb.drop(['id','host_name','last_review'], axis=1, inplace=True)
# Fill 'reviews_per_month' as Nan with 0
airbnb.fillna({'reviews_per_month':0}, inplace=True)
airbnb.head()
```

```python
airbnb['reviews_per_month'].isnull().sum()
```

## outlier observation and processing

```python
airbnb['minimum_nights'].plot.box()
```

```python
# Replace the column greater than 365 with 365
airbnb.loc[airbnb['minimum_nights'] > 365, 'minimum_nights'] = 365
```

```python
# Delete rows with price less than or equal to 0
airbnb.drop(airbnb[airbnb.price<=0].index, inplace=True)
```
